#### Pandas-DataFrame And Series
Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning. It provides two primary data structures: Series and DataFrame. A Series is a one-dimensional array-like object, while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes (rows and columns).

```python
import pandas as pd
```

```python
## Series
##A Pandas Series is a one-dimensional array-like object that can hold any data type. It is similar to a column in a table.

import pandas as pd
data=[1,2,3,4,5]
series=pd.Series(data)
print("Series \n",series)
print(type(series))
```

```python
## Create a Series from dictionary
data={'a':1,'b':2,'c':3}
series_dict=pd.Series(data)
print(series_dict)
```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data,index=index)
```

```python
## Dataframe
## create a Dataframe from a dictionary oof list

data={
    'Name':['Krish','John','Jack'],
    'Age':[25,30,45],
    'City':['Bangalore','New York','Florida']
}
df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
## Create a Data frame From a List of Dictionaries

data=[
    {'Name':'Krish','Age':32,'City':'Bangalore'},
    {'Name':'John','Age':34,'City':'Bangalore'},
    {'Name':'Bappy','Age':32,'City':'Bangalore'},
    {'Name':'JAck','Age':32,'City':'Bangalore'}
    
]
df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
''' 
| Series        | DataFrame            |
| ------------- | -------------------- |
| 1-D           | 2-D                  |
| Single column | Multiple columns     |
| Has index     | Has rows and columns |
| `pd.Series()` | `pd.DataFrame()`     |

Series = 1D labeled array (like a single Excel column).
Default index = 0,1,2,...
Dictionary keys become the index of a Series.
Custom index can be given using index=[...].
DataFrame = Collection of Series arranged as rows and columns (like an Excel sheet).
pd.Series() → creates a Series.
pd.DataFrame() → creates a DataFrame

'''
```

```python
'''
Suppose
df.index = [100,200,300]


Now
df.loc[100]
returns
Krish
because 100 is the label.

But

df.iloc[0]
also returns Krish because it's the 0th position.

So
loc → label/index based
iloc → integer position based

This difference becomes important when your index is not 0,1,2,....
'''
```

```python
## Accessing a specified element 
df.at[2,'Age']
```

```python
''' 
Syntax
df.at[row_label, column_name]

Returns one value.
It is optimized for fast scalar access.

'''
```

```python
df.at[2,'Name']
```

```python
## Accessing a specified element using iat
df.iat[2,2]

''' 
Row 2
Column 2

'''
```

```python
'''  

pd.read_csv() → Reads a CSV file into a DataFrame.
head(n) → First n rows (default = 5).
tail(n) → Last n rows (default = 5).
df["Column"] → Returns a Series (single column).
loc → Label-based indexing (row_label, column_name).
iloc → Integer position-based indexing (row_position, column_position).
at → Fast access to a single value using label + column name.
iat → Fast access to a single value using row position + column position.

| Function    | Uses                 | Based On            | Returns          |
| ----------- | -------------------- | ------------------- | ---------------- |
| `head(n)`   | First n rows         | -                   | DataFrame        |
| `tail(n)`   | Last n rows          | -                   | DataFrame        |
| `df["col"]` | One column           | Column name         | Series           |
| `loc`       | Row/column selection | Label               | Series/DataFrame |
| `iloc`      | Row/column selection | Position            | Series/DataFrame |
| `at`        | Single value         | Label + Column name | Scalar           |
| `iat`       | Single value         | Position + Position | Scalar           |
| `drop()`    | Remove rows/columns  | Label               | New DataFrame    |

'''
```

```python
df.drop(0)

'''  
Row 0 is removed in the returned DataFrame.
Very Important
The original DataFrame is not modified


still gives
Krish
John
Jack
because
drop() returns a new DataFrame.

To permanently remove:
df = df.drop(0)
or
df.drop(0, inplace=True)
'''
```

```python
### Data MAnipulation with Dataframe
df
```

```python
## Adding a column
df['Salary']=[50000,60000,70000] #The number of values must match the number of rows
df
```

```python
## Remove a column
df.drop('Salary',axis=1,inplace=True)

''' 
Why axis=1?
Pandas has two axes.
axis=0 → Rows
axis=1 → Columns

'''
```

```python
df
```

```python
## Add age to the column
df['Age']=df['Age']+1
df
```

```python
df.drop(0,inplace=True)

'''  
Notice

Index is still
1
2

It does not become
0
1

To reset the index:
df.reset_index(drop=True, inplace=True)

'''
```
