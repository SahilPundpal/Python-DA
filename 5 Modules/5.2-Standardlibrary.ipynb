```python
## csv

import csv

with open('example.csv',mode='w',newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['Krish',32])

with open('example.csv',mode='w',newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['Sahil',22])

with open('example.csv',mode='r') as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)


''' 

What happens?
The second block also uses:
mode='w'

which overwrites the file.
So after the second block, the file contains only:

name,age
Sahil,22

That's exactly why your example.csv shows:

'''
```

```python
## datetime
from datetime import datetime,timedelta

now=datetime.now()
print(now)

yesterday=now-timedelta(days=1)
## yesterday=now-timedelta(month=1)-> gives error 

print(yesterday)
```

```python
## time
import time
print(time.time())
time.sleep(5)
print(time.time())
```

```python
## Regular expresiion
import re

pattern=r'\d+'
text='There are 123 apples 456'
match=re.search(pattern,text)
print(match.group())
```
