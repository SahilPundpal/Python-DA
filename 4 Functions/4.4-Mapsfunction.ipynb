#### The map() Function in Python
The map() function applies a given function to all items in an input list (or any other iterable) and returns a map object (an iterator). This is particularly useful for transforming data in a list comprehensively.

```python
def square(x):
    return x*x

square(10)
```

```python
numbers=[1,2,3,4,5,6,7,8]

print(list(map(square,numbers)))
## print(map(square,numbers)) this won't work cause
 
'''
 What is map()?
map() returns a map object, which is an iterator.
Think of it as a machine that knows how to generate values one by one, but hasn't generated them all yet.
'''

print(tuple(map(square, numbers)))
```

```python
## Lambda function with map
numbers=[1,2,3,4,5,6,7,8]
list(map(lambda x:x*x,numbers))
```

```python
### MAp multiple iterables

numbers1=[1,2,3]
numbers2=[4,5,6]

added_numbers=list(map(lambda x,y:x+y,numbers1,numbers2))
print(added_numbers)


### MAp multiple iterables

numbers1=[1,2,3]
numbers2=[4,5,6]
numbers3=[9,8,7]
added_numbers=list(map(lambda x,y,z:x+y+z,numbers1,numbers2,numbers3))
print(added_numbers)


numbers1=[1,2,3]
numbers2=[4,5,6]
numbers3=[9,8]
added_numbers=list(map(lambda x,y,z:x+y+z,numbers1,numbers2,numbers3))
print(added_numbers)

##map() stops when the shortest iterable is exhausted.
```

```python
## map() to convert a list of strings to integers
# Use map to convert strings to integers
str_numbers = ['1', '2', '3', '4', '5']
int_numbers = list(map(int, str_numbers))

str_numbers1 = ['1''2', '2', '3', '4', '5']
int_numbers1=list(map(int, str_numbers1))
print(int_numbers)  # Output: [1, 2, 3, 4, 5]
print(int_numbers1)  # Output: [12, 2, 3, 4, 5]
```

```python
words=['apple','banana','cherry']
upper_word=list(map(str.upper,words))
print(upper_word)

words=['apple','banana','cherry']
upper_word=list(map(str.lower,words))
print(upper_word)
```

```python
def get_name(person):
    return person['name']

people=[
    {'name':'Krish','age':32},
    {'name':'Jack','age':33}
]
print(list(map(get_name,people)))

def get_name2(person):
    return person['name']
people1={
    "itm1":{'name':'Krish','age':32},
    "itm2":{'name':'Jack','age':33}
}
print(list(map(get_name2,people1.values())))

'''  
map(func, dict) → passes dictionary KEYS.
map(func, dict.values()) → passes dictionary VALUES.
map(func, dict)          -> passes dictionary KEYS
map(func, dict.keys())   -> passes dictionary KEYS
map(func, dict.values()) -> passes dictionary VALUES
map(func, dict.items())  -> passes (key, value) tuples
List → map() passes each element.

Dictionary → map() passes KEYS by default.

Need dictionary values? Always use .values().
Need both key and value? Use .items().


'''
```
